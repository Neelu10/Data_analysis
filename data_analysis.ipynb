```python
ax=sns.countplot(x='Fuel_Price',data=df)
```

```python
ax=sns.countplot(x='Fuel_Price',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Date'],as_index=False)['Weekly_Sales'].sum().sort_values(by='Weekly_Sales',ascending=False)
sns.barplot( x ='Date',y='Weekly_Sales',data=sales_gen)
```

```python
df.columns
```

```python
ax=sns.countplot(data=df, x='Holiday_Flag')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
storeq=df.groupby(['Temperature'],as_index=False)['Date'].sum().sort_values(by='Date',ascending=False).head(10)

sns.set(rc={'figure.figsize':(100,200)})
sns.barplot(data=storeq,x='Temperature',y='Date')
```
